```python

from pyspark.sql.types import StructType, StructField, StringType, IntegerType

from pyspark.sql.window import Window

employee_data = [
    (1, "Alice", "IT", 100000),
    (2, "Bob", "IT", 120000),
    (3, "Charlie", "IT", 120000),
    (4, "David", "HR", 90000),
    (5, "Eva", "HR", 95000),
    (6, "Frank", "HR", 95000),
    (7, "Grace", "Finance", 110000),
    (8, "Hank", "Finance", 105000),
    (9, "Ivy", "Finance", 110000),
    (10, "Jack", "Finance", 110000)
]

input_schema=StructType([
    StructField("emp_id", IntegerType(),True),
    StructField("name", StringType(),True),
    StructField("department", StringType(),True),
    StructField("salary", IntegerType(),True)
    
])

#columns = ["emp_id", "name", "department", "salary"]

# create data frame
df_employee=spark.createDataFrame(employee_data,input_schema)

# display(df_employee)
df_employee.show()
df_employee.printSchema()

#create temporary view
df_employee.createOrReplaceTempView("employee")

```
