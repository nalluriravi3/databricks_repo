```python
import requests
import pandas as pd

url = "https://httpbin.org/basic-auth/ravi/password123"

response = requests.get(
    url,
    auth=("ravi", "password123")
)

response.raise_for_status()

print(response.json())

pdf = pd.json_normalize(response.json())

df = spark.createDataFrame(pdf)

display(df)
```
