```python
#create gold_df
from pyspark.sql.functions import trim,col
gold_df=spark.table("workspace.silver_new.silver_customer")

error_message=""

try:
    gold_df=gold_df.filter(col("salary")>70000)\
        .drop("ingestion_timestamp",)\
        .withColumn("customer_id",trim(col("customer_id")).cast("int"))\
        .orderBy(col("customer_id").asc())

    display(gold_df)

    #write into gold table
    gold_df.write.format("delta")\
        .option("mergeSchema", "true") \
        .mode("overwrite") \
        .saveAsTable("workspace.gold_new.gold_customer")

except Exception as e:
    error_message=str(e)

if error_message:
    dbutils.notebook.exit(f"FAILED: {error_message}")
else:
    dbutils.notebook.exit("SUCCESS")

```
