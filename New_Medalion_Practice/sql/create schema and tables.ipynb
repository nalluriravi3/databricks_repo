## Create Schema

## Create Audit Table

```python
%sql
CREATE TABLE IF NOT EXISTS workspace.audit_new.audit_log (
    pipeline_name STRING,
    layer_name STRING,
    table_name STRING,
    load_start_time TIMESTAMP,
    load_end_time TIMESTAMP,
    records_read BIGINT,
    records_written BIGINT,
    records_rejected BIGINT,
    status STRING,
    error_message STRING
)
USING DELTA;
```

## Create Quarantine Table

```python
%sql
CREATE TABLE IF NOT EXISTS workspace.quarantine_new.customer_quarantine (
    customer_id STRING,
    customer_name STRING,
    contact STRING,
    email STRING,
    salary INT,
    rejected_reason STRING,
    rejected_timestamp TIMESTAMP
)
USING DELTA;
```

## Create PIpeline Audit Table

```python
%sql
CREATE TABLE IF NOT EXISTS workspace.audit_new.pipeline_audit
(
    batch_id STRING,
    pipeline_name STRING,
    notebook_name STRING,
    start_time TIMESTAMP,
    end_time TIMESTAMP,
    duration_seconds BIGINT,
    status STRING,
    error_message STRING
)
USING DELTA;
```
