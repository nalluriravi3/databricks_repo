```python
# check if file exists in a particular location
dbutils.widgets.text("folder_path","")
dbutils.widgets.text("file_name","")

folder_path=dbutils.widgets.get("folder_path")
file_name=dbutils.widgets.get("file_name")
                               
print(f"folder name is:{folder_path}")
print(f"file name is:{file_name}")

def file_exists(folder_path,file_name):
    try:
        files=dbutils.fs.ls(folder_path)
        for path in files:
            if path.name==file_name:
                return True
                

        return False
    except:
        return False


if file_exists(folder_path,file_name):
    print("File exists. Executing notebook...")
    output= "Found"
    
    
   
else:
    print("File not found. Notebook execution skipped.")
    output= "File not found"

#set the value output to be used in other note books
dbutils.jobs.taskValues.set("file_exist",output)
print(output)
```
