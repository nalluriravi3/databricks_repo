```python
class Employee:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def display(self):
        print(f"Name:{self.name} and Age:{self.age}")
        
emp=Employee("Ravi",43)
emp.display()
```

```python
# simple inhetance
class Vehicle:
    def __init__(self,brand,year):
        self.brand=brand
        self.year=year
        print("in parent")
    def display(self):
        print(f"inside parent and vechicle brand is:{self.brand}")

class Car(Vehicle):
    def __init__(self,brand,year,model):
        super().__init__(brand,year)
        self.model=model
        print("in child")
        
    def show(self):
        print(f"inside child and vechicle brand is:{self.brand} ")
        
        
car1=Car("BMW",2010,"X5")
car1.show()

        

```

```python
emp = {
    "id": 101,
    "name": "John"
}

skills = ["Python", "SQL"]

try:
    print(emp["salary"])      # KeyError
    print(skills[5])          # IndexError
except KeyError as e:
    print("Missing key:", e)
except IndexError as e:
    print("Invalid index:", e)
```

```python
class Vehicle:    
    def display(self):
        print("Vehicle method executed")


class Car(Vehicle):
    def show(self):
        super().display()      # Parent method
        raise Exception("Error in child method")


car = Car()

try:
    car.show()
except Exception as e:
    print("Caught:", e)
```
