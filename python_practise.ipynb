```python
#program to return true if paris are possible
def can_form_pairs(A):
    evens=[e for e in A if e%2==0]
    odds=[o for o in A if o%2!=0 ]

    print(evens)
    print(odds)

    # evens=[2,4,6]
    # odds=[7,3,17]

    if len(evens)!=len(odds):
        return False
    else:
        pairs=[(e,o) for e,o in zip(evens,odds)]
        return pairs
    

nums = [2, 7, 4, 6, 3, 17]
can_form_pairs(nums)


```

```python
nums = [2, 7, 4, 6, 3, 9]

evens=[e for e in nums if e%2==0]
odds=[o for o in nums if o%2!=0 ]

if len(evens)!=len(odds):
    print("pairing is not possible")
else:
    pairs=[(e,o) for e,o in zip(evens,odds)]
for a,b in pairs:
    assert a%2==0
    assert (a+b)%2!=0
print(pairs)
```

```python
# Write a function to print prime numbers in the given range
def prime(start,end):
    for n in range(start,end+1):
        if n>1:
            for i in range(2,n):
                if n%i==0:
                    break
            else:
                print(n)
prime(2,5)
```

```python
# write a program to find requency of elements in a list
list1=[1,2,3,4,5,1]
freq={}
for i in list1:
    if i in freq:
        freq[i]+=1
    else:
        freq[i]=1
    

print(freq)

# print most repeated element 
most_repeated_key=0
value_count=0
for k,v in freq.items():
    if v>value_count:
        most_repeated_key=k
        value_count=v
print(f"{most_repeated_key}:{value_count}")
```

```python
#find missing number in list
from pyspark.sql.functions import col,min,max
data=[(1,),(3,),(5,),(7,),(9,),(10,)]

#create data frame
df=spark.createDataFrame(data,["id"])
display(df)

#find min and max numbers
min_id=df.select(min("id")).collect()[0][0]
max_id=df.select(max("id")).collect()[0][0]

print(min_id)
print(max_id)

#create a data frame wth all numbers
df1=spark.range(min_id,max_id)
display(df1)

#perform join
df2=df1.join(df,df1.id==df.id,"leftanti")
display(df2)
```

```python
# practise joins  to display number of records
data1=[1,1,1,2,2,3,3,4]
data2=[1,1,2,3,None,None]

df1=spark.createDataFrame(data1,["id"])
df2=spark.createDataFrame(data2,["id"])

# innder join
df3=df1.join(df2,df1.id==df2.id,"inner")
print(df3.count())
# left outer join
df3=df1.join(df2,df1.id==df2.id,"leftouter")
print(df3.count())
# right outer join
df3=df1.join(df2,df1.id==df2.id,"rightouter")
print(df3.count())
# full outer join
df3=df1.join(df2,df1.id==df2.id,"outer")
print(df3.count())
#cartesion join
df3=df1.crossJoin(df2)
print(df3.count())
#left semi
df3=df1.join(df2,df1.id==df2.id,"leftsemi")
print(df3.count())
#left anti
df3=df1.join(df2,df1.id==df2.id,"leftanti")
print(df3.count())
```
