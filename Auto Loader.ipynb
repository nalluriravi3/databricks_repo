## Auto Loader

```python
from pyspark.sql.functions import *
df=spark.readStream.format("cloudFiles") \
  .option("cloudFiles.format", "csv") \
  .option("cloudFiles.schemaLocation", "/Volumes/workspace/default/autosink/schema location") \
  .load("/Volumes/workspace/default/autosource/") \
 
```

```python
from pyspark.sql.functions import *
df.writeStream\
    .format("parquet")\
    .option("checkpointlocation","/Volumes/workspace/default/autosink/schema location") \
    .trigger(availableNow=True) \
    .start("/Volumes/workspace/default/autosink/data")
```
