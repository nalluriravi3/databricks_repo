```python
#implement SCD1
delta_table_obj = DeltaTable.forName(spark, "target_table")

delta_table_obj.alias("tgt") \
.merge(
    source = src_df.alias("src"),
    condition = col("tgt.id") == col("src.id")
) \
.whenMatchedUpdate(
    set = {
        "name": col("src.name"),
        "salary": col("src.salary")
    }
) \
.whenNotMatchedInsert(
    values = {
        "id": col("src.id"),
        "name": col("src.name"),
        "salary": col("src.salary")
    }
) \
.execute()
```
