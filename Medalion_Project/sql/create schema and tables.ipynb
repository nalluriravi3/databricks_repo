## Creat audit table

```python
%sql
CREATE TABLE IF NOT EXISTS workspace.audit.audit_log (
    pipeline_name STRING,
    layer_name STRING,
    table_name STRING,
    load_start_time TIMESTAMP,
    load_end_time TIMESTAMP,
    records_read BIGINT,
    records_written BIGINT,
    records_rejected BIGINT,
    status STRING,
    error_message STRING
)
USING DELTA;
```

Create quarantine table

```python
%sql
CREATE TABLE IF NOT EXISTS workspace.quarantine.customer_quarantine (
    customer_id STRING,
    customer_name STRING,
    contact STRING,
    email STRING,
    salary INT,
    rejected_reason STRING,
    rejected_timestamp TIMESTAMP
)
USING DELTA;
```
