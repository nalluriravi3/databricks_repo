```python
from pyspark.sql.functions import col,trim
gold_df=spark.table("workspace.silver.customer_silver")
gold_df=gold_df.filter(col("salary")>70000)\
    .drop("ingestion_timestamp",)\
    .withColumn("customer_id",trim(col("customer_id")).cast("int"))\
    .orderBy(col("customer_id").asc())

#write into gold table
gold_df.write.format("delta")\
       .option("mergeSchema", "true") \
       .mode("overwrite") \
       .saveAsTable("workspace.gold.customer_gold")
```
